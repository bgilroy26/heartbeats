```python
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
sig = np.fromfile('ptbdb/patient001/s0010_re.dat', dtype=np.int16)[13211:22800:12]
```

```python
t = np.linspace(0, 1, 1000, False)
plt.plot(sig, c='blue')
sos = scipy.signal.butter(10, [.005,.05], output='sos', btype='bandpass')
filtered = scipy.signal.sosfiltfilt(sos, sig)
plt.plot(filtered, c='orange')
plt.show()
```

```python
sigstart105, sigend105 = 1091, 1382400
t105 = np.linspace(sigstart105, sigend105, int((sigend105-sigstart105)/12), False)
sig105 = np.fromfile('ptbdb/patient105/s0303lre.dat', dtype=np.int16)[::12][91:]
```

```python
import sys
plt.figure(figsize=(20,10))
plt.plot(t105, sig105, c='blue')
sos105 = scipy.signal.butter(10, [.005, .07], output='sos', btype='bandpass')
filtered105 = scipy.signal.sosfiltfilt(sos105, sig105)
polyfit105 = np.polynomial.polynomial.Polynomial.fit(t105, filtered105, 8)
polyval105 = polyfit105.linspace(int((sigend105-sigstart105)/12), [0, sigend105])
final105 = filtered105 - polyval105[1]
plt.plot(t105, final105, c='orange')
peaks, _ = scipy.signal.find_peaks(final105, distance=720)

plt.plot(t105[peaks], final105[peaks], "x")
plt.show()
```
